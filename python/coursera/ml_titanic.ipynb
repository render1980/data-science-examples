```python
import pandas as pnd
import math
```

```python
data = pnd.read_csv('titanic.csv', index_col='PassengerId')
```

```python
data
```

```python
def writeToFile(f, val):
    with open(f, "a") as f_:
        f_.write(val)
```

```python
# Кол-во мужчин на корабле
male_cnt = len(data[(data.Sex.eq("male"))])
# Кол-во женщин на корабле
female_cnt = len(data[(data.Sex.eq("female"))])

sex_cnt = str(male_cnt) + ' ' + str(female_cnt)
#writeToFile('1.txt', sex_cnt)
sex_cnt
```

```python
# Доля выживших
survived_len = len(data[(data.Survived.eq(1))])
survived_proportion = format(float(survived_len) / len(data) * 100, '2.0f')
survived_proportion
#writeToFile('2.txt', survived_proportion)
survived_proportion
```

```python
# Доля пассажиров 1-го класс
firstclass_len = len(data[data.Pclass.eq(1)])
firstclass_proportion = format(float(firstclass_len) / len(data) * 100, '2.0f')
firstclass_proportion
#writeToFile('3.txt', firstclass_proportion)
firstclass_proportion
```

```python
# Возраст: среднее и медиана
age_mean = format(data.Age.mean(), '2.2f')
age_median = format(data.Age.median(), '2.0f')
mean_median = str(age_mean) + ' ' + str(age_median)
#writeToFile('4.txt', mean_median)
mean_median
```
